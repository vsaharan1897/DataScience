```python
#Cell to Run:

def CheckThisWeekFiles(file_path):
    AirlineCodeTester=f'{file_path[0:2]}'
    ThisWeekTester='This Week'
    LastWeekTester='Last Week'
    file_path=pd.read_csv(file_path, skiprows=5)
    InitalIndex=file_path.index[file_path['Travel Date'] == 'Marketing Airlines'].tolist()
    DataLoaded=file_path.index[file_path['Travel Date']=='Use Data Loaded'].tolist()
    TravelPeriod=[DataLoaded[0]+4]
    AirlineCodeResult=file_path['Airline Code'][2]
    DataLoadedResult=file_path['Flights'][DataLoaded[0]]
    TravelPeriodResult=file_path['Flights'][TravelPeriod[0]]
    if AirlineCodeResult == AirlineCodeTester:
        pass
    #     print("Correct Airline")
    else:
        print(f'{AirlineCodeTester}:')
        print("WRONG AIRLINE")
    if DataLoadedResult == ThisWeekTester:
        pass
    #     print("Correct Week")
    else:
        print(f'{AirlineCodeTester}:')
        print("WRONG WEEK")
    if '2024' in TravelPeriodResult:
        pass
    #     print("Correct Travel Period")
    else:
        print(f'{AirlineCodeTester}:')
        print("WRONG TRAVEL PERIOD")

def CheckLastWeekFiles(file_path):
    AirlineCodeTester=f'{file_path[0:2]}'
    LastWeekTester='Last Week'
    file_path=pd.read_csv(file_path, skiprows=5)
    InitalIndex=file_path.index[file_path['Travel Date'] == 'Marketing Airlines'].tolist()
    DataLoaded=file_path.index[file_path['Travel Date']=='Use Data Loaded'].tolist()
    TravelPeriod=[DataLoaded[0]+4]
    AirlineCodeResult=file_path['Airline Code'][2]
    DataLoadedResult=file_path['Flights'][DataLoaded[0]]
    TravelPeriodResult=file_path['Flights'][TravelPeriod[0]]
    if AirlineCodeResult == AirlineCodeTester:
        pass
    #     print("Correct Airline")
    else:
        print(f'{AirlineCodeTester}:')
        print("WRONG AIRLINE")
    if DataLoadedResult == LastWeekTester:
        pass
    #     print("Correct Week")
    else:
        print(f'{AirlineCodeTester}:')
        print("WRONG WEEK")
    if '2024' in TravelPeriodResult:
        pass
    #     print("Correct Travel Period")
    else:
        print(f'{AirlineCodeTester}:')
        print("WRONG TRAVEL PERIOD")

def CheckLastYearFiles(file_path):
    AirlineCodeTester=f'{file_path[0:2]}'
    file_path=pd.read_csv(file_path, skiprows=5)
    InitalIndex=file_path.index[file_path['Travel Date'] == 'Marketing Airlines'].tolist()
    DataLoaded=file_path.index[file_path['Travel Date']=='Use Data Loaded'].tolist()
    TravelPeriod=[DataLoaded[0]+4]
    AirlineCodeResult=file_path['Airline Code'][2]
    DataLoadedResult=file_path['Flights'][DataLoaded[0]]
    TravelPeriodResult=file_path['Flights'][TravelPeriod[0]]
    if AirlineCodeResult == AirlineCodeTester:
        pass
    #     print("Correct Airline")
    else:
        print(f'{AirlineCodeTester}:')
        print("WRONG AIRLINE")
    if '2022' in TravelPeriodResult:
        pass
    #     print("Correct Travel Period")
    else:
        print(f'{AirlineCodeTester}:')
        print("WRONG TRAVEL PERIOD")
```

```python
warnings.filterwarnings("ignore")
ThisWeekFiles=['AATWPullData.csv','UATWPullData.csv','ASTWPullData.csv','DLTWPullData.csv','B6TWPullData.csv','WNTWPullData.csv'
          ,'F9TWPullData.csv', 'G4TWPullData.csv','MXTWPullData.csv','XPTWPullData.csv','SYTWPullData.csv','NKTWPullData.csv']
for TWfile in ThisWeekFiles:
    TWresult=CheckThisWeekFiles(TWfile)
    TWresult
```
